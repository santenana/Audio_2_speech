```python
import requests

def obtener_palabras_ofensivas():
    fuentes = {
        "es": "https://github.com/jfreddypuentes/spanlp/tree/master/spanlp/dataset",
        "en": "https://github.com/Jenyay/Obscene-Words-List/blob/master/languages/en",
    }
    palabras = {}
    for idioma, url in fuentes.items():
        try:
            respuesta = requests.get(url)
            if respuesta.status_code == 200:
                lista = [p.strip().lower() for p in respuesta.text.split("\n") if p.strip()]
                palabras[idioma] = lista
            else:
                palabras[idioma] = []
        except Exception as e:
            print(f"Error obteniendo lista para {idioma}: {e}")
            palabras[idioma] = []

    palabras_ofensivas_es = palabras.get("es", [])
    palabras_ofensivas_en = palabras.get("en", [])

    return palabras_ofensivas_es, palabras_ofensivas_en

# Ejemplo de uso:
es, en = obtener_palabras_ofensivas()
print(es)
print(en)
```
